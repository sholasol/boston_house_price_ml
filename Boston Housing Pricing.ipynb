## Standarding Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
# Standardize X_train

X_train = scaler.fit_transform(X_train)
```

## Model Training -  Implement Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
# Train the model (Using training data set)
regression.fit(X_train, y_train)
```

```python
# Print the coefficients and intercept

print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# On which parameter the model has been trained
regression.get_params()
```

### Prediction with Test Data

```python
# for the X_test, y_test is the true value
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
# Use performance metrics to evaluate model performance
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(f'MAE: {mean_absolute_error(y_test, reg_pred)}')
print(f'mse: {mean_squared_error(y_test, reg_pred)}')
print(f'RMS: {np.sqrt(mean_squared_error(y_test, reg_pred))}')
```

```python
# Performance Interpretation 

# MAE measures the average absolute difference between the predicted values and the actual values. 
# It is calculated by taking the average of the absolute differences between the predicted and actual values 
# for each data point. In your case, the MAE is 0.5272474538305953. A lower MAE indicates that the model's 
# predictions are closer to the actual values on average.


# Mean Squared Error (MSE):
# MSE is another metric that calculates the average of the squared differences between the predicted and 
# actual values. It gives more weight to larger errors due to the squaring operation. In your case, 
# the MSE is 0.5305677824766752. Like MAE, a lower MSE value indicates that the model's predictions are 
# closer to the actual values.

# Root Mean Squared Error (RMSE):
# RMSE is the square root of MSE and provides a more interpretable metric in the same units as the target variable. 
# It represents the average magnitude of the error between the predicted and actual values. In your case, 
# the RMSE is 0.7284008391515452. Similar to MAE and MSE, a lower RMSE value suggests that the model's predictions 
# have smaller errors on average.
```

# R square and Adjusted R Square

##### Formula

##### R^2 is the Coefficient of Determinaion
##### SSR is the Sum of residual squarea
##### SST is the Total Sum of Squared

##### R^2 = 1 - SSR/SST


```python
from sklearn.metrics import r2_score

score = r2_score(y_test, reg_pred)
print(score)
```

```python
# the model perform above average at 59%
```

#### Adjusted R Squared

##### R2 is the R Square of the model
##### n is the number of Observation 
##### k is the number of predictor variables

##### R^2 = 1 - [(1-R2)*(n-1)/(n-k-1)]

```python
# Display the adjusted R-squared

1 - (1 - score)*(len(y_test) - 1 )/(len(y_test) - X_test.shape[1] - 1)
```
